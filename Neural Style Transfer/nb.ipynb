```python
style_images_path = [
    "./style images/emotional-joshua-miels.jpg",
    "./style images/van-gogh.jpg"
]
```

```python
from loss import NST_Loss
from torch.optim import Adam, lr_scheduler

device = torch.device("mps")

tensor_loader = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(
        (0.485, 0.456, 0.406),
        (0.229, 0.224, 0.225)
    ),
])
tensor_unloader = transforms.Compose([
    transforms.Normalize(
        (-2.12, -2.04, -1.80),
        (4.37, 4.46, 4.44)
    ),
    transforms.ToPILImage()
])

def load_image(image_path: str, ):
    """Load an image and convert it to a torch tensor."""
    image = Image.open(image_path)

    img_shape = image.size
    scale_factor = 400 / max(img_shape)
    # scale_factor = 400 / max(img_shape)
    new_img_shape = np.array(img_shape) * scale_factor
    image = image.resize(new_img_shape.astype(int), Image.Resampling.LANCZOS)

    image = tensor_loader(image).unsqueeze(0)
    return image.to(device)

style_images = [
    load_image(path_)
    for path_
    in style_images_path
]
content_image = load_image("./content images/The-Boy.jpeg")
generated_image = content_image.clone().requires_grad_(True)

learning_rate = 0.03

NST_Network = Network(style_images, content_image)
loss_fxn = NST_Loss(NST_Network.style_features, NST_Network.content_features)
optimizer = Adam([generated_image], lr=learning_rate)
scheduler = lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.01)

loss_data = {}

# for epoch in range(100):
#     generated_features = NST_Network(generated_image)
    
#     loss, misc_loss_data = loss_fxn(generated_features)
#     loss_data[epoch] = misc_loss_data
    
#     optimizer.zero_grad()
#     loss.backward()
#     optimizer.step()
# tensor_unloader(generated_image.squeeze(0))
epoch = 0
```

```python
for epoch in range(epoch, epoch+10):
    generated_features = NST_Network(generated_image)
    
    loss, misc_loss_data = loss_fxn(generated_features)
    loss_data[epoch] = misc_loss_data
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
cur_loss_data = loss_data[epoch]    
print(f"Epoch: {epoch+1}, Loss: {cur_loss_data['total_loss']:.2f}, Style Loss: {cur_loss_data['style_loss']:.2f}, Content Loss: {cur_loss_data['content_loss']:.2f}")
tensor_unloader(generated_image.squeeze(0))
```
